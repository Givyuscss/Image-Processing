```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pywt

image = cv2.imread('tmp.jpg')

gray_image = 1/3 * image[:,:,0] + 1/3 * image[:,:,1] + 1/3 * image[:,:,2]
m,n = gray_image.shape
gray_image = gray_image.astype('uint8')

coeffs = pywt.dwt2(gray_image, 'haar')
cA, (cH, cV, cD) = coeffs

plt.imshow(gray_image,'gray')
```

```python
cA = np.where(cA > 250, 1.5 * cA,cA)
cA = np.where(cA <= 150, 0.5 * cA,cA)
```

```python
img = pywt.idwt2((cA, (cH, cV, cD)),'haar').astype('uint8')
#cv2.imwrite('res_tmp.jpg',img.astype('uint8'))
# 

plt.imshow(img,'gray')
plt.title('result')
plt.show()
```
